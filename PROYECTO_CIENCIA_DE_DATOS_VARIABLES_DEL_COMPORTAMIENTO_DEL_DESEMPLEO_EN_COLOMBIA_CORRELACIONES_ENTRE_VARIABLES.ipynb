```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Cargar los datos desde el archivo
df = pd.read_excel("Evolucion_Indicadores_Economicos_Colombia_1990_2025.xlsx")

# Calcular la regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(df["Año"], df["Tasa de Desempleo (%)"])

# Crear el modelo de regresión
df["Predicción"] = intercept + slope * df["Año"]

# Configurar el estilo de Seaborn
sns.set_theme(style="whitegrid")

# Crear el gráfico de dispersión con puntos conectados
plt.figure(figsize=(12, 6))
plt.scatter(df["Año"], df["Tasa de Desempleo (%)"], c=df["Tasa de Desempleo (%)"], cmap="coolwarm", alpha=0.8, edgecolors="black", label="Datos reales")
plt.plot(df["Año"], df["Tasa de Desempleo (%)"], linestyle="-", color="black", alpha=0.6, marker="o", label="Tendencia observada")
plt.plot(df["Año"], df["Predicción"], linestyle="--", color="red", alpha=0.8, label="Regresión lineal")

# Agregar los coeficientes de correlación y determinación
plt.text(1992, max(df["Tasa de Desempleo (%)"]) - 2, f"Coef. de Correlación: {r_value:.2f}", fontsize=12, color="red")
plt.text(1992, max(df["Tasa de Desempleo (%)"]) - 3, f"Coef. de Determinación (R²): {r_value**2:.2f}", fontsize=12, color="red")

# Mejoras en el diseño
plt.xlabel("Año", fontsize=12)
plt.ylabel("Tasa de Desempleo (%)", fontsize=12)
plt.title("Año vs Tasa de Desempleo (%) en Colombia (1990-2025)", fontsize=14, fontweight="bold")
plt.colorbar(label="Tasa de Desempleo (%)")  # Muestra la intensidad del desempleo
plt.xticks(rotation=45)  # Inclina los años para mejor lectura
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
