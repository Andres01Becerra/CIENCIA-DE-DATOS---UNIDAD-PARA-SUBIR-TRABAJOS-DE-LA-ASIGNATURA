```python
from google.colab import files
import pandas as pd

# Subir archivo manualmente desde la computadora
uploaded = files.upload()

# Obtener el nombre del archivo
file_name = list(uploaded.keys())[0]

# Cargar el archivo en un DataFrame
df = pd.read_excel(file_name)

# Ver las primeras 5 filas
df.head()
```

```python
df.info()
df.describe()
df.head(100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df['Edad'], bins=20, kde=True)
plt.title('Distribución de la Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.savefig("histograma_edad.png")
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(y=df['Departamento'], order=df['Departamento'].value_counts().index)
plt.title('Distribución de Empleados por Departamento')
plt.xlabel('Cantidad de Empleados')
plt.ylabel('Departamento')
plt.savefig("barras_departamento.png")
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Salario'], y=df['Puntaje_Desempeño'])
plt.title('Relación entre Salario y Puntaje de Desempeño')
plt.xlabel('Salario ($)')
plt.ylabel('Puntaje de Desempeño')
plt.savefig("dispersion_salario_puntaje.png")
plt.show()
```

```python
df.to_csv("datos_procesados.csv", index=False)
from google.colab import files
files.download("datos_procesados.csv")
files.download("histograma_edad.png")
files.download("barras_Especie.png.")
files.download("dispersion_ID_Cliente_puntaje.png")
```

```python
from google.colab import files
import pandas as pd

# Subir archivo manualmente desde la computadora
uploaded = files.upload()

# Obtener el nombre del archivo
file_name = list(uploaded.keys())[0]

# Cargar el archivo en un DataFrame
df = pd.read_excel(file_name)

# Ver las primeras 5 filas
df.head()
```

```python
df.info()
df.describe()
df.head(30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df['Edad'], bins=20, kde=True)
plt.title('Distribución de la Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.savefig("histograma_edad.png")
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(y=df['Especie'], order=df['Especie'].value_counts().index)
plt.title('Distribución de Especies por cliente')
plt.xlabel('Cantidad de Especie')
plt.ylabel('Especie')
plt.savefig("barras_Especie.png")
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['ID_Cliente'], y=df['Nombre'])
plt.title('Relación entre ID_Cliente y Nombre')
plt.xlabel('ID_Cliente (#)')
plt.ylabel('Nombre')
plt.savefig("dispersion_ID_Cliente_puntaje.png")
plt.show()
```

```python
df.to_csv("datos_procesados.csv", index=False)
from google.colab import files
files.download("datos_procesados.csv")
files.download("histograma_edad.png")
files.download("barras_Especie.png")
files.download("dispersion_ID_Cliente_puntaje.png")
```
