##*Aplicabilidad*

1. **Series temporales:** Las series temporales se usan para analizar datos dependientes del tiempo, como los precios históricos de acciones.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos simulados de precios de acciones
data = {'Fecha': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
        'Precio': [50 + x*0.2 for x in range(365)]}
df = pd.DataFrame(data)

# Convertir columna Fecha a tipo datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Graficar precios
plt.plot(df['Fecha'], df['Precio'])
plt.title('Evolución del precio de las acciones')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.show()
```

2. **Análisis financiero:** Determinar el rendimiento acumulado de una inversión utilizando datos temporales.

```python
# Calcular el rendimiento acumulado
df['Rendimiento'] = (df['Precio'] / df['Precio'].iloc[0]) - 1

# Graficar rendimiento acumulado
plt.plot(df['Fecha'], df['Rendimiento'])
plt.title('Rendimiento acumulado')
plt.xlabel('Fecha')
plt.ylabel('Rendimiento (%)')
plt.show()
```

3. **Seguimiento de sensores:** Procesar datos generados por sensores en intervalos de tiempo regulares, como temperatura ambiental.

```python
import pandas as pd
import numpy as np

# Simulación de datos de sensores
sensor_data = {
    'Timestamp': pd.date_range(start='2023-01-01 00:00:00', periods=1000, freq='H'),
    'Temperatura': np.random.normal(25, 2, 1000)  # Generación de datos simulados con distribución normal
}
df_sensores = pd.DataFrame(sensor_data)

# Filtrar por un rango de fechas
inicio = pd.to_datetime('2023-01-15 00:00:00')  # Convertir las fechas a objetos datetime
fin = pd.to_datetime('2023-01-20 23:59:59')
df_filtrado = df_sensores[(df_sensores['Timestamp'] >= inicio) & (df_sensores['Timestamp'] <= fin)]

# Resumen de datos
resumen = df_filtrado.describe()  # Cálculo de estadísticas descriptivas para el rango filtrado
print("Resumen de datos filtrados:")
print(resumen)
```

4. **Datos de clima (otro contexto):** Analizar temperaturas promedio por mes.

```python
# Agregar columna de mes y calcular promedio mensual
df_sensores['Mes'] = df_sensores['Timestamp'].dt.month
promedio_mensual = df_sensores.groupby('Mes')['Temperatura'].mean()

print(promedio_mensual)
```

```python
import pandas as pd
from datetime import datetime

# Crear un rango de fechas
fechas = pd.date_range(start='2023-01-01', periods=10, freq='D')

# Crear el DataFrame
df = pd.DataFrame({'fechas': fechas})
df['año'] = df['fechas'].dt.year
df['mes'] = df['fechas'].dt.month
df['día'] = df['fechas'].dt.day
df['día_semana'] = df['fechas'].dt.day_name()

# Calcular cuántos días han pasado desde una fecha específica
fecha_referencia = datetime(2023, 1, 5)  # Fecha de referencia
df['días_desde_referencia'] = (df['fechas'] - fecha_referencia).dt.days

# Filtrar los datos por fechas (ejemplo: obtener fechas después del 2023-01-05)
filtro_fecha = '2023-01-05'
df_filtrado = df[df['fechas'] > filtro_fecha]

# Mostrar resultados
print("DataFrame original con columna 'días_desde_referencia':")
print(df)
print("\nDatos filtrados (fechas posteriores a 2023-01-05):")
print(df_filtrado)
```

#**3. Jerarquía de Fechas.**

Organizar y agrupar datos por componentes temporales implica estructurar la información en función de períodos definidos, como años, trimestres, meses, semanas o días. Esto permite identificar patrones y tendencias específicas dentro de cada intervalo de tiempo.

---

Las herramientas para agrupar datos temporales incluyen:

- Año: Permite observar tendencias anuales, ideal para estudios macroeconómicos.

- Trimestre: Útil para análisis financieros y planificación estratégica.

- Mes: Identifica variaciones mensuales, como estacionalidades.

- Semana: Rastrea comportamientos semanales, valioso en marketing y logística.

- Día o Día de la Semana: Detecta fluctuaciones diarias o patrones recurrentes específicos.

En Python, la librería pandas facilita esta agrupación. Por ejemplo:

```python
import pandas as pd

# Simulación de datos
data = {'Fecha': pd.date_range(start='2023-01-01', periods=100, freq='D'),
        'Ventas': [100 + x * 2 for x in range(100)]}
df = pd.DataFrame(data)

# Agrupar por mes y sumar las ventas
df['Mes'] = df['Fecha'].dt.month
ventas_mensuales = df.groupby('Mes')['Ventas'].sum()

# Agrupar por día de la semana
df['Día_semana'] = df['Fecha'].dt.day_name()
ventas_por_día = df.groupby('Día_semana')['Ventas'].mean()

print("Ventas mensuales:\n", ventas_mensuales)
print("Promedio de ventas por día de la semana:\n", ventas_por_día)
```

##*Aplicabilidad*

1. **Dashboards:** Crear un resumen visual de ventas diarias y mensuales.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulación de datos
data = {'Fecha': pd.date_range(start='2023-01-01', periods=100, freq='D'),
        'Ventas': [100 + x*2 for x in range(100)]}
df = pd.DataFrame(data)

# Agregar columnas temporales
df['Mes'] = df['Fecha'].dt.month

# Resumen mensual para el dashboard
ventas_mensuales = df.groupby('Mes')['Ventas'].sum()

# Gráficos del dashboard
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(df['Fecha'], df['Ventas'], label='Ventas Diarias')
plt.title('Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Ventas')

plt.subplot(1, 2, 2)
ventas_mensuales.plot(kind='bar')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.tight_layout()
plt.show()
```

2. **Reportes Mensuales:** Obtener estadísticas descriptivas de ventas mensuales.

```python
# Calcular estadísticas mensuales
stats_mensuales = df.groupby('Mes')['Ventas'].describe()
print("Estadísticas mensuales:\n", stats_mensuales)
```

 3. **Pronósticos:** Predecir valores futuros con una tendencia simple.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Crear un modelo de regresión lineal
df['Días'] = range(len(df))  # Agregar número de días como variable independiente
X = df[['Días']]
y = df['Ventas']
modelo = LinearRegression()
modelo.fit(X, y)

# Pronóstico para los próximos 10 días
futuro = pd.DataFrame({'Días': range(len(df), len(df)+10)})
futuro['Pronóstico'] = modelo.predict(futuro[['Días']])

# Visualizar el pronóstico
plt.plot(df['Días'], df['Ventas'], label='Ventas Reales')
plt.plot(futuro['Días'], futuro['Pronóstico'], label='Pronóstico', linestyle='--')
plt.title('Pronóstico de Ventas')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Supongamos un DataFrame con ventas diarias
df = pd.DataFrame({
    'fecha': pd.date_range(start='2023-01-01', periods=90, freq='D'),
    'ventas': np.random.randint(100, 1000, size=90)
})

# Imprimir las fechas correctamente
print(df['fecha'])

# Agrupación por mes
df['mes'] = df['fecha'].dt.to_period('M')
ventas_mensuales = df.groupby('mes')['ventas'].sum()

# Imprimir el DataFrame original y las ventas mensuales
print(df)
print(ventas_mensuales)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creación del DataFrame con ventas diarias
df = pd.DataFrame({
    'fecha': pd.date_range(start='2023-01-01', periods=90, freq='D'),
    'ventas': np.random.randint(100, 1000, size=90)
})

# Agrupación por trimestre y cálculo de estadísticas descriptivas
df['trimestre'] = df['fecha'].dt.to_period('Q')  # Agregamos columna de trimestre
estadisticas_trimestrales = df.groupby('trimestre')['ventas'].describe()  # Estadísticas descriptivas
print("Estadísticas descriptivas por trimestre:")
print(estadisticas_trimestrales)

# Agrupación por mes y cálculo del promedio
df['mes'] = df['fecha'].dt.to_period('M')  # Agregamos columna de mes
ventas_promedio_mensual = df.groupby('mes')['ventas'].mean()  # Promedio de ventas por mes

# Visualización de las ventas promedio por mes
plt.figure(figsize=(8, 6))
ventas_promedio_mensual.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Ventas Promedio por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas Promedio')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
